```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = sns.load_dataset('penguins')
```

```python
sns.regplot(x='bill_length_mm', y='bill_depth_mm', data=df)
plt.title('Linear Regression Plot')
plt.show()
```

```python
# Multiple Linear Regression Plot
sns.lmplot(x='bill_length_mm', y='bill_depth_mm', hue='species', data=df)
plt.title('Multiple Linear Regression Plot')
plt.show()
```

```python
# Residual Plot
sns.residplot(x='bill_length_mm', y='bill_depth_mm', data=df)
plt.title('Residual Plot')
plt.show()
```

```python
# Histogram
sns.histplot(df['bill_length_mm'], kde=False)
plt.title('Histogram')
plt.show()
```

```python
# KDE Plot
sns.kdeplot(df['bill_length_mm'].dropna(), shade=True)
plt.title('Kernel Density Estimate Plot')
plt.show()
```

```python
# Distribution Plot
sns.histplot(df['bill_length_mm'].dropna(), kde=True)
plt.title('Distribution Plot')
plt.show()
```

```python
# Box Plot
sns.boxplot(x='species', y='bill_length_mm', data=df)
plt.title('Box Plot')
plt.show()
```

```python
# Violin Plot
sns.violinplot(x='species', y='bill_length_mm', data=df)
plt.title('Violin Plot')
plt.show()
```

```python
# Pair Plot
sns.pairplot(df.dropna())
plt.title('Pair Plot')
plt.show()
```

```python
# Joint Plot
sns.jointplot(x='bill_length_mm', y='bill_depth_mm', data=df, kind='scatter')
plt.title('Joint Plot')
plt.show()
```

```python
# Facet Grid
g = sns.FacetGrid(df, col='species')
g.map(plt.hist, 'bill_length_mm')
plt.suptitle('Facet Grid')
plt.show()
```
