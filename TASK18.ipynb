## Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset("tips")

# Create scatter plot
sns.scatterplot(data=data, x="total_bill", y="tip")
 
sns.set_style("whitegrid")
plt.title("Scatter Plot of Total Bill vs Tip")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

```python
# Create line plot
sns.lineplot(data=data, x="total_bill", y="tip")
```

```python
# Create bar plot
sns.barplot(data=data, x="day", y="total_bill")

 
sns.set_style("whitegrid")
plt.title("Bar Plot of Total Bill by Day")
plt.xlabel("Day")
plt.ylabel("Total Bill")
plt.show()
```

```python
# Create count plot
sns.countplot(data=data, x="day")
 
sns.set_style("darkgrid")
plt.title("Count Plot of Days")
plt.xlabel("Day")
plt.ylabel("Count")
plt.show()
```

## Customization

```python
import seaborn as sns
import matplotlib.pyplot as plt

 
data = sns.load_dataset("tips")

 
sns.set_style("whitegrid")
sns.set_palette("bright")

 
sns.scatterplot(data=data, x="total_bill", y="tip", hue="day", style="time", size="size", palette="bright")

 
plt.title("Scatter Plot of Total Bill vs Tip")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```
