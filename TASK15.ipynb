```python
# Importing libraries
import pandas as pd
import numpy as np

# Initializing the data
data = {
    'Names': ['Ali', 'Bilal', 'Hammad', 'Abdullah'],
    'Age': ['21-25', '21-25', '21-25', '21-25'],
    'Gender': ['Male', 'Male', 'Male', 'Male'],
    'Mobile Phone': ['Yes', 'Yes', 'Yes', 'Yes'],
    'Mobile Operating System': ['Android', 'Android', 'iOS', 'Android'],
    'Mobile phone use for education': ['Sometimes', 'Sometimes', 'Sometimes', 'Frequently'],
    'Mobile phone activities': ['Social Media', 'Social Media', 'All of these', 'All of these'],
    'Helpful for studying': ['Yes', 'Yes', 'Yes', 'Yes'],
    'Educational Apps': ['Educational Videos', 'Educational Videos', 'Educational Videos', 'Educational Videos'],
    'Daily usages': ['4-6 hours', '4-6 hours', '4-6 hours', '2-4 hours'],
    'Performance impact': ['Agree', 'Neutral', 'Strongly agree', 'Strongly agree'],
    'Usage distraction': ['During Exams', 'During Exams', 'Not Distracting', 'During Class Lectures'],
    'Attention span': ['Yes', 'Yes', 'No', 'No'],
    'Useful features': ['Camera', 'Notes Taking App', 'Camera', 'Internet Access'],
    'Health Risks': ['Yes', 'Yes', 'Only Partially', ''],
    'Beneficial subject': ['Accounting', 'Browsing Material', 'Research', ''],
    'Usage symptoms': ['Headache', 'All of these', 'Sometimes', 'Never'],
    'Symptom frequency': ['Never', 'Sometimes', 'Sometimes', ''],
    'Health precautions': ['Using Blue light filter', 'Taking Break during prolonged use', 'None of Above', 'Limiting Screen Time'],
    'Health rating': ['Excellent', 'Good', 'Excellent', 'Excellent']
}

# Creating the DataFrame
df = pd.DataFrame(data)
```

```python
# Creating a sample second DataFrame
data2 = {
    'Names': ['Ali', 'Bilal', 'Hammad', 'Abdullah'],
    'Score': [85, 79, 92, 88]
}

df2 = pd.DataFrame(data2)

# Merge on a single key
merged_single_key = pd.merge(df, df2, on='Names')
print("Merged DataFrame on single key:\n", merged_single_key)
```

```python
# Adding more columns to df2 for multiple key merge
data2 = {
    'Names': ['Ali', 'Bilal', 'Hammad', 'Abdullah'],
    'Age': ['21-25', '21-25', '21-25', '21-25'],
    'Score': [85, 79, 92, 88]
}

df2 = pd.DataFrame(data2)

# Merge on multiple keys
merged_multiple_keys = pd.merge(df, df2, on=['Names', 'Age'])
print("Merged DataFrame on multiple keys:\n", merged_multiple_keys)
```

```python
# Outer join
outer_join = pd.merge(df, df2, on='Names', how='outer')

# Inner join
inner_join = pd.merge(df, df2, on='Names', how='inner')

# Left join
left_join = pd.merge(df, df2, on='Names', how='left')

# Right join
right_join = pd.merge(df, df2, on='Names', how='right')

print("Outer Join:\n", outer_join)
print("\nInner Join:\n", inner_join)
print("\nLeft Join:\n", left_join)
print("\nRight Join:\n", right_join)
```

```python
# Concatenate along rows
concatenated_rows = pd.concat([df, df2], ignore_index=True)
print("Concatenated along rows:\n", concatenated_rows)
```

```python
# Concatenate along columns
concatenated_columns = pd.concat([df, df2], axis=1)
print("Concatenated along columns:\n", concatenated_columns)
```

```python
# Concatenate a list of DataFrames
list_of_dfs = [df, df2]
concatenated_list = pd.concat(list_of_dfs)
print("Concatenated list of DataFrames:\n", concatenated_list)
```

```python
# Reshape using melt
melted_data = pd.melt(df, id_vars=['Names', 'Age', 'Gender'], var_name='Variable', value_name='Value')
print("Melted DataFrame:\n", melted_data)
```

```python
# Convert 'Health rating' to numeric  
df['Health rating'] = pd.to_numeric(df['Health rating'], errors='coerce')

# Create a pivot table
pivot_table = df.pivot_table(index='Names', columns='Age', values='Health rating', aggfunc='mean')
print("Pivot Table:\n", pivot_table)
```

```python
# Grouping and aggregating
grouped_data = df.groupby(['Age', 'Gender']).agg({'Daily usages': 'sum', 'Health rating': 'mean'})
print("Grouped and aggregated data:\n", grouped_data)
```

```python
# Convert 'Daily usages' to numeric  
df['Daily usages'] = pd.to_numeric(df['Daily usages'], errors='coerce')

#  custom aggregation function
def custom_agg(x):
    return x.max() - x.min()

 
custom_aggregated = df.groupby('Mobile Operating System').agg({
    'Daily usages': custom_agg,
    'Health rating': 'mean'
})
print("Custom aggregated data:\n", custom_aggregated)
```
