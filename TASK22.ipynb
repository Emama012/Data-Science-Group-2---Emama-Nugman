```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate sample data
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

plt.scatter(X, y)
plt.title("Sample Data")
plt.xlabel("X")
plt.ylabel("y")
plt.show()
```

```python
# Initialize parameters (weights and bias)
def initialize_parameters():
    W = np.random.randn(1, 1)
    b = np.random.randn(1)
    return W, b

# Hypothesis function
def hypothesis(X, W, b):
    return np.dot(X, W) + b
    # Checking initialized parameters
W, b = initialize_parameters()
print("Initial Weight:", W)
print("Initial Bias:", b)

```

```python
# Mean Squared Error (MSE) loss function
def compute_loss(X, y, W, b):
    m = len(y)
    y_pred = hypothesis(X, W, b)
    loss = (1 / (2 * m)) * np.sum((y_pred - y) ** 2)
    return loss


# Checking MSE calculation
initial_loss = compute_loss(X, y, W, b)
print("Initial Loss:", initial_loss)

```

```python
# Gradient Descent to update weights and bias
def gradient_descent(X, y, W, b, learning_rate, num_iterations):
    m = len(y)
    loss_history = []
    
    for i in range(num_iterations):
        y_pred = hypothesis(X, W, b)
        
        dW = (1 / m) * np.dot(X.T, (y_pred - y))
        db = (1 / m) * np.sum(y_pred - y)
        
        W -= learning_rate * dW
        b -= learning_rate * db
        
        loss = compute_loss(X, y, W, b)
        loss_history.append(loss)
        
        if i % 100 == 0:
            print(f"Iteration {i}: Loss = {loss}")
    
    return W, b, loss_history


# Training the model and displaying final parameters
W, b, loss_history = gradient_descent(X, y, W, b, learning_rate, num_iterations)
print("Final Weight:", W)
print("Final Bias:", b)
print("Final Loss:", loss_history[-1])

```

```python
# Training the Linear Regression model from scratch
W, b = initialize_parameters()
learning_rate = 0.01
num_iterations = 1000

W, b, loss_history = gradient_descent(X, y, W, b, learning_rate, num_iterations)
```

```python
# Plotting the loss history
plt.plot(range(num_iterations), loss_history)
plt.title("Loss History")
plt.xlabel("Iterations")
plt.ylabel("Loss")
plt.show()
```

```python
# Plotting the regression line
plt.scatter(X, y)
plt.plot(X, hypothesis(X, W, b), color='red')
plt.title("Linear Regression Fit")
plt.xlabel("X")
plt.ylabel("y")
plt.show()
```

```python
# Making predictions
X_new = np.array([[0], [2]])
y_pred = hypothesis(X_new, W, b)
print("Predictions:", y_pred)
```
